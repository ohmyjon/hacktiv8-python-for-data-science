```python
from tkinter import ttk, filedialog
from openpyxl.workbook.protection import WorkbookProtection
from tkinter import messagebox
import tkinter as tk
import openpyxl
import pandas as pd
import xlwings as xw
import os
import threading

# Function template

# Filled loyalty template from data region
def entry_automation(sheet, region, lr, month, calc):
    
    if region == 'R5': 
        for row in range(19, lr + 1):
        
            if calc == 'm1':
                # April
                # Segmentation Customer
                sheet['G' + str(row)].value = segmentData(data=month, sheet=sheet, row=row)
                # AQUA-600ML
                sheet['BH' + str(row)].value = sumData(data=month, sheet=sheet, row=row, product='AQUA-600ML')  
                # AQUA-1500ML
                sheet['BI' + str(row)].value = sumData(data=month, sheet=sheet, row=row, product='AQUA-1500ML')
                # AQUA-220ML
                sheet['BJ' + str(row)].value = sumData(data=month, sheet=sheet, row=row, product='AQUA-220ML')
                # AQUA-330ML
                sheet['BK' + str(row)].value = sumData(data=month, sheet=sheet, row=row, product='AQUA-330ML')
                # AQUA-5GL
                sheet['BL' + str(row)].value = sumData(data=month, sheet=sheet, row=row, product='AQUA-5GL')
                # VIT-220ML
                sheet['BM' + str(row)].value = sumData(data=month, sheet=sheet, row=row, product='VIT-220ML')
                # VIT-330ML
                sheet['BN' + str(row)].value = sumData(data=month, sheet=sheet, row=row, product='VIT-330ML')
                # VIT-550ML
                sheet['BO' + str(row)].value = sumData(data=month, sheet=sheet, row=row, product='VIT-550ML')
                # VIT-1500ML
                sheet['BP' + str(row)].value = sumData(data=month, sheet=sheet, row=row, product='VIT-1500ML')
                # VIT-5GL
                sheet['BQ' + str(row)].value = sumData(data=month, sheet=sheet, row=row, product='VIT-5GL')
                # MIZONE
                sheet['BR' + str(row)].value = sumData(data=month, sheet=sheet, row=row, product='MIZONE')

      
            elif calc == 'm2':
                # Mei
                # Segmentation Customer
                sheet['H' + str(row)].value = segmentData(data=month, sheet=sheet, row=row)
                # AQUA-600ML
                sheet['BX' + str(row)].value = sumData(data=month, sheet=sheet, row=row, product='AQUA-600ML')  
                # AQUA-1500ML
                sheet['BY' + str(row)].value = sumData(data=month, sheet=sheet, row=row, product='AQUA-1500ML')
                # AQUA-220ML
                sheet['BZ' + str(row)].value = sumData(data=month, sheet=sheet, row=row, product='AQUA-220ML')
                # AQUA-330ML
                sheet['CA' + str(row)].value = sumData(data=month, sheet=sheet, row=row, product='AQUA-330ML')
                # AQUA-5GL
                sheet['CB' + str(row)].value = sumData(data=month, sheet=sheet, row=row, product='AQUA-5GL')
                # VIT-220ML
                sheet['CC' + str(row)].value = sumData(data=month, sheet=sheet, row=row, product='VIT-220ML')
                # VIT-330ML
                sheet['CD' + str(row)].value = sumData(data=month, sheet=sheet, row=row, product='VIT-330ML')
                # VIT-550ML
                sheet['CE' + str(row)].value = sumData(data=month, sheet=sheet, row=row, product='VIT-550ML')
                # VIT-1500ML
                sheet['CF' + str(row)].value = sumData(data=month, sheet=sheet, row=row, product='VIT-1500ML')
                # VIT-5GL
                sheet['CG' + str(row)].value = sumData(data=month, sheet=sheet, row=row, product='VIT-5GL')
                # MIZONE
                sheet['CH' + str(row)].value = sumData(data=month, sheet=sheet, row=row, product='MIZONE')
      
            elif calc == 'm3':
                # Juni
                # Segmentation Customer
                sheet['I' + str(row)].value = segmentData(data=month, sheet=sheet, row=row)
                # AQUA-600ML
                sheet['CN' + str(row)].value = sumData(data=month, sheet=sheet, row=row, product='AQUA-600ML')  
                # AQUA-1500ML
                sheet['CO' + str(row)].value = sumData(data=month, sheet=sheet, row=row, product='AQUA-1500ML')
                # AQUA-220ML
                sheet['CP' + str(row)].value = sumData(data=month, sheet=sheet, row=row, product='AQUA-220ML') 
                # AQUA-330ML
                sheet['CQ' + str(row)].value = sumData(data=month, sheet=sheet, row=row, product='AQUA-330ML') 
                # AQUA-5GL
                sheet['CR' + str(row)].value = sumData(data=month, sheet=sheet, row=row, product='AQUA-5GL') 
                # VIT-220ML
                sheet['CS' + str(row)].value = sumData(data=month, sheet=sheet, row=row, product='VIT-220ML') 
                # VIT-330ML
                sheet['CT' + str(row)].value = sumData(data=month, sheet=sheet, row=row, product='VIT-330ML') 
                # VIT-550ML
                sheet['CU' + str(row)].value = sumData(data=month, sheet=sheet, row=row, product='VIT-550ML') 
                # VIT-1500ML
                sheet['CV' + str(row)].value = sumData(data=month, sheet=sheet, row=row, product='VIT-1500ML') 
                # VIT-5GL
                sheet['CW' + str(row)].value = sumData(data=month, sheet=sheet, row=row, product='VIT-5GL') 
                # MIZONE
                sheet['CX' + str(row)].value = sumData(data=month, sheet=sheet, row=row, product='MIZONE') 
    
            else:
                print('Error: Bulan tidak ditemukan')

    elif region in ('R1', 'R2', 'R3', 'R4', 'R6', 'RBN'):     
        for row in range(19, lr + 1):
            
            if calc == 'm1':
                # April
                # Segmentation Customer
                sheet['G' + str(row)].value = segmentData(data=month, sheet=sheet, row=row)
                # AQUA-600ML
                sheet['BH' + str(row)].value = sumData(data=month, sheet=sheet, row=row, product='AQUA-600ML')  
                # AQUA-1500ML
                sheet['BI' + str(row)].value = sumData(data=month, sheet=sheet, row=row, product='AQUA-1500ML')
                # AQUA-220ML
                sheet['BJ' + str(row)].value = 0
                # AQUA-330ML
                sheet['BK' + str(row)].value = 0
                # AQUA-5GL
                sheet['BL' + str(row)].value = 0
                # VIT-220ML
                sheet['BM' + str(row)].value = 0
                # VIT-330ML
                sheet['BN' + str(row)].value = 0
                # VIT-550ML
                sheet['BO' + str(row)].value = 0
                # VIT-1500ML
                sheet['BP' + str(row)].value = 0
                # VIT-5GL
                sheet['BQ' + str(row)].value = 0
                # MIZONE
                sheet['BR' + str(row)].value = 0
            
            elif calc == 'm2':
                # Mei
                # Segmentation Customer
                sheet['H' + str(row)].value = segmentData(data=month, sheet=sheet, row=row)
                # AQUA-600ML
                sheet['BX' + str(row)].value = sumData(data=month, sheet=sheet, row=row, product='AQUA-600ML')  
                # AQUA-1500ML
                sheet['BY' + str(row)].value = sumData(data=month, sheet=sheet, row=row, product='AQUA-1500ML')
                # AQUA-220ML
                sheet['BZ' + str(row)].value = 0
                # AQUA-330ML
                sheet['CA' + str(row)].value = 0
                # AQUA-5GL
                sheet['CB' + str(row)].value = 0
                # VIT-220ML
                sheet['CC' + str(row)].value = 0
                # VIT-330ML
                sheet['CD' + str(row)].value = 0
                # VIT-550ML
                sheet['CE' + str(row)].value = 0
                # VIT-1500ML
                sheet['CF' + str(row)].value = 0
                # VIT-5GL
                sheet['CG' + str(row)].value = 0
                # MIZONE
                sheet['CH' + str(row)].value = 0
                
            elif calc == 'm3':
                # Juni
                # Segmentation Customer
                sheet['I' + str(row)].value = segmentData(data=month, sheet=sheet, row=row)
                # AQUA-600ML
                sheet['CN' + str(row)].value = sumData(data=month, sheet=sheet, row=row, product='AQUA-600ML')  
                # AQUA-1500ML
                sheet['CO' + str(row)].value = sumData(data=month, sheet=sheet, row=row, product='AQUA-1500ML')
                # AQUA-220ML
                sheet['CP' + str(row)].value = 0
                # AQUA-330ML
                sheet['CQ' + str(row)].value = 0
                # AQUA-5GL
                sheet['CR' + str(row)].value = 0
                # VIT-220ML
                sheet['CS' + str(row)].value = 0
                # VIT-330ML
                sheet['CT' + str(row)].value = 0
                # VIT-550ML
                sheet['CU' + str(row)].value = 0
                # VIT-1500ML
                sheet['CV' + str(row)].value = 0
                # VIT-5GL
                sheet['CW' + str(row)].value = 0
                # MIZONE
                sheet['CX' + str(row)].value = 0
            
            else:
                print('Error: Bulan tidak ditemukan')

    else:
        print('Error: Region tidak ditemukan')

    return sheet

# Last row for loyalty template
def lastRow(sheet):
    last_row = sheet.max_row
    while sheet.cell(column=1, row=last_row).value is None and last_row > 0:
        last_row -= 1
    return last_row

# Sum data based on Customer_Id
def sumData(data, sheet, row, product):
    _subset = data[(data['Customer_Id'] == sheet['D' + str(row)].value) & (data['brand_prod'] == product)]
    _subset = _subset['In Box'].sum()
    return _subset

# Segment data based on subsegmentDet
def segmentData(data, sheet, row):
    _subset = data[(data['Customer_Id'] == sheet['D' + str(row)].value)].tail(1)['subsegmentDet'].squeeze()
    if type(_subset) == pd.Series:
        _subset = None
    return _subset

# -------------------------------------------------------------------------------------------------------------------------------

# Main Program
cwd = os.getcwd()

LARGE_FONT = ("Arial", 12)
MEDIUM_FONT = ("Arial", 11)
REGULAR_FONT = ("Arial", 10)

text_z = ["Select Loyalty Template file", "Select M1 file", "Select M2 file", "Select M3 file"]

window = tk.Tk()
window.title('Loyalty Calculation Program')

def click(x): 
    c = tk.filedialog.askopenfilename(initialdir=cwd, title="Select file", filetypes=(("Excel file", "*.xlsx"), ("All files", "*.*")))
    file[x].insert(tk.END, c)

    
# def click_region():
#     global region
#     region = tmp.get()

tmp = tk.StringVar()

# ttk.Label(window, text="file load", font=LARGE_FONT, background="white").grid(row=0, column=0, padx=20, pady=10, sticky="w")
ttk.Label(window, text="Enter Region, ex: R6, RBN", background="white").grid(row=0, column=0, columnspan=3, padx=10, pady=5, sticky="w")
input_region = ttk.Entry(window, textvariable=tmp, width=60, background="gray").grid(row=1, column=0, columnspan=3, padx=10, sticky="ew")

file = []
for i in range(len(text_z)): 
    frame = tk.Frame(window)
    frame.grid(row=i+2, column=0, sticky="nsew")
    ttk.Label(frame, text=text_z[i], background="white").grid(row=0, column=0, columnspan=3, padx=10, pady=2, sticky="w")
    file.append(ttk.Entry(frame, width=60, background="gray"))
    file[i].grid(row=1, column=0, columnspan=3, padx=10, sticky="ew")
    ttk.Button(frame, text="Search", width=10, command=lambda x=i: click(x)).grid(row=1, column=3, padx=5, sticky="w")

def main_program():
    
    print('Menjalankan program..')
    region = (tmp.get()).upper()
    wb = openpyxl.load_workbook(file[0].get())
    ws = wb.active
    last_row = lastRow(ws)
    
    for i in range(1,4):
        f = file[i].get()
        if f != '':
            print(f'Mengkalkulasi M{i}')
            m = pd.read_excel(f)
            ws = entry_automation(sheet=ws, region=region, lr=last_row, month=m, calc=f'm{i}')
            del m
    
    path = f'{cwd}\\{region}'
    os.makedirs(path, exist_ok=True)
    wb.save(f'{cwd}\\{region}\\{region} Loyalty Calculation.xlsx')
    wb.close()
    
    distri = []
    for row in range(19, last_row + 1):
        # Filtering Distributor
        if ws['C' + str(row)].value != None:
            distri.append(ws['C' + str(row)].value)
    
    xw.App().visible = False

    filtering = xw.Book(f'{cwd}\\{region}\\{region} Loyalty Calculation.xlsx')
    filtering.sheets[0].api.Protect(Password='6fedbc1245446b9c920d0dd9b1bfd29b', AllowFormattingCells=True, AllowFormattingColumns=True, AllowFormattingRows=True, AllowFiltering=True)
    filtering.save(f'{cwd}\\{region}\\{region} Loyalty Calculation.xlsx')
    filtering.sheets[0].api.Unprotect(Password='6fedbc1245446b9c920d0dd9b1bfd29b')
    
    print('Memulai filtering distributor')
    for d in set(distri):
        filtering.sheets[0].api.Range(f'A18:ED{last_row}').AutoFilter(3, d)
        filtering.sheets[0].api.Protect(Password='6fedbc1245446b9c920d0dd9b1bfd29b', AllowFormattingCells=True, AllowFormattingColumns=True, AllowFormattingRows=True, AllowFiltering=True)
        if ('/') in d:
            d = d.replace('/', '\N{FULLWIDTH SOLIDUS}')
        filtering.save(f'{region}\\{region} Loyalty Calculation - {d}.xlsx')
        filtering.sheets[0].api.Unprotect(Password='6fedbc1245446b9c920d0dd9b1bfd29b')
        filtering.sheets[0].api.AutoFilterMode = False

    filtering.close()
    
    print('Selesai')
    messagebox.showinfo('Loyalty Calculation Program', 'Program is done')

tk.Button(window, text='Mulai', command=threading.Thread(target=main_program).start).grid()

window.mainloop()
```
